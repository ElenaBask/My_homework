```python
def get_experience(arg):
    if arg is np.nan or arg == 'Не указано':
        return np.nan
    arg_spl = arg.split()[2:6]
    years = ['лет', 'год', 'года']
    months = ['месяц', 'месяца', 'месяцев']
    counter = 0
    for i, n in enumerate(arg_spl):
        if n in years:
            counter += int(arg_spl[i-1]) * 12
        if n in months:
            counter += int(arg_spl[i-1])
    return counter
        
    
```

```python
def get_sity(arg): #функция по преобразованию городов
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    big_cities = ['Москва', 'Санкт-Петербург']
    sity = arg.split()[0]
    if sity in big_cities:
        return sity
    elif sity in million_cities:
        return 'город-миллионник'
    else: return 'другие'

def ready_to_move(arg): # функция по анализу готовности к переезду
    if ('не готова к переезду') in arg or ('не готов к переезду' in arg): # если человек не отказывается, значит он либо готов, либо хочет переехать
        return False
    else: return True

def ready_to_trip(arg):
    if 'командировка' in arg:
        if 'не готов к командировкам' in arg: # считаем, что готов,  даже если указано "готов к редким"
            return False
        else: return True
    else: return True
```

```python
HHData['Город'] = HHData['Город, переезд, командировки'].apply(get_sity)
HHData['Готовность к переезду'] = HHData['Город, переезд, командировки'].apply(ready_to_move)
HHData['Готовность к командировкам'] = HHData['Город, переезд, командировки'].apply(ready_to_trip)
```

```python
def get_currency(currency):
    
    standard_iso = {"грн":"UAH", "USD":"USD", "EUR":"EUR", "белруб":"BYN", "KGS":"KGS", "сум":"UZS", "AZN":"AZN", "KZT":"KZT"}
    currency_info = currency.split(' ')[1].replace('.', '')
    
    if currency_info == 'руб':
        return 'RUB'
    else:
        return standard_iso[currency_info]


def get_payment(payment):
    
    payment_info = payment.split(" ")
    return float(payment_info[0])
    

HHData["Обновление резюме"] = pd.to_datetime(HHData['Обновление резюме']).dt.date
HHData["Наименование валюты"] = HHData["ЗП"].apply(get_currency)
HHData["Cумма ЗП"] = HHData["ЗП"].apply(get_payment)

data_currency = pd.read_csv("data/ExchangeRates.csv", sep=",")
data_currency["date"] = pd.to_datetime(data_currency["date"], dayfirst=True).dt.date

HHData = HHData.merge(data_currency, how="left", left_on=["Обновление резюме", "Наименование валюты"], right_on=["date", "currency"])

HHData["close"] = HHData["close"].fillna(1)
HHData["proportion"] = HHData["proportion"].fillna(1)

HHData["ЗП (руб)"] = HHData["Cумма ЗП"]*HHData["close"] / HHData["proportion"]
round(HHData["ЗП (руб)"].median() /1000)
```
